# **Task 1: Conceptual Understanding**

**1. What is the difference between "Love" and "love" in NLP?**

In NLP, "Love" and "love" are considered different words if the text is not normalized. This is because computers treat uppercase and lowercase letters differently.

However, during preprocessing, we usually convert all text to lowercase. After this step, both "Love" and "love" become the same word, which helps in reducing duplication and improving consistency in analysis.


**2. What happens if stopwords are not removed?**

If stopwords (like is, the, and, in) are not removed, they can create noise in the data. These words appear very frequently but usually do not carry important meaning.

due to this:
* The model may focus on unimportant words
* Processing becomes slower due to more data
* Accuracy of tasks like classification or sentiment analysis may decrease

**3. Provide two real-world scenarios where removing stopwords can be harmful.**

1] Sentiment Analysis**

Words like "not" are very important.

    Example:
    "I am happy"
    "I am not happy"
If "not" is removed, both sentences will look the same, which gives wrong results.

2] Search Engines / Question Answering

Stopwords can change the meaning of a query.

    Example:
    "Flights to Dubai"
    "Flights from Dubai"
Words like to and from are important here. Removing them can lead to incorrect search results.

**4. Explain the difference between stemming and lemmatization.**

Stemming and lemmatization are both used to reduce words to their base form, but they work differently:

* Stemming : It simply cuts off prefixes or suffixes without considering the meaning.

    Example:
    running → run,
    studies → studi


* Lemmatization : It converts a word to its correct base form using vocabulary and grammar rules.It is more accurate but slightly slower.

    Example:
    running → run,
    better → good




# **Task 2: Build Advanced Preprocessing Function**

```python
import re

def preprocess_text(text):
    try:
        # Error handling
        if not text or text.strip() == "":
            return [], ""

        # Convert to lowercase
        text = text.lower()

        # Remove URLs
        text = re.sub(r'http\S+|www\S+', '', text)

        # Remove email patterns
        text = re.sub(r'\S+@\S+', '', text)

        # Remove numbers
        text = re.sub(r'\d+', '', text)

        # Handle repeated characters (tooo → to)
        text = re.sub(r'(.)\1{2,}', r'\1', text)

        # Remove emojis & special character
        text = re.sub(r'[^a-z\s]', '', text)

        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        # Tokenization
        tokens = text.split()

        # Remove short tokens (≤2) except "no", "not"
        cleaned_tokens = [
            word for word in tokens
            if len(word) > 2 or word in ["no", "not"]
        ]

        cleaned_sentence = " ".join(cleaned_tokens)

        return cleaned_tokens, cleaned_sentence

    except Exception as e:
        print("Error:", e)
        return [], ""
```

# **Task 3: Stress Testing**

```python
sentences = [
     "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://google.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

print("--- Stress Testing Results ---")

for text in sentences:
    tokens, cleaned = preprocess_text(text)

    print("\nOriginal Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)

    results.append((tokens, cleaned))
```

# **Task 4: Token Analytics**


```python
def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(t)
     for t in tokens) / total_tokens

    return total_tokens, unique_tokens, round(avg_length, 2)


print("\n--- Token Analytics ---")

for i, (tokens, _) in enumerate(results):
    total, unique, avg = token_analytics(tokens)

    print(f"\nSentence {i+1}:")
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", avg)
```

# **Task 5: Frequency Analysis**

```python
from collections import Counter

all_tokens = []

for tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("\n--- Frequency Analysis ---")

print("\nTop 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

# **Task 6: Build Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(sentences)

print("\nAll Tokens:", result["tokens"])
print("\nClean Sentences:", result["clean_sentences"])
```
